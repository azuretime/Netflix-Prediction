## Data Preprocessing and Feature Engineering

```python
#### Feature embedding from Image preprocessing
## get 'poster_index.csv' and 'poster_features.csv' from "image.ipynb" file

poster_index = np.loadtxt('poster_index.csv',delimiter=',')
print(poster_index.shape)

poster_features = np.loadtxt('poster_features.csv',delimiter=',')
print(poster_features.shape)

np.set_printoptions(suppress=True)

index_to_keep = [int(i) for i in poster_index]
filtered_data = df.drop(index=df.index.difference(index_to_keep))
filtered_data.shape

poster_features = pd.DataFrame(poster_features)
poster_features.shape

filtered_data.reset_index(drop=True, inplace=True)
poster_features.reset_index(drop=True, inplace=True)

new_columns = {}
for i in range(len(poster_features.columns)):
    new_columns[poster_features.columns[i]] = "poster_" + str(i+1)

poster_features = poster_features.rename(columns=new_columns)
```

```python
data_merge = pd.concat([filtered_data,poster_features],axis=1)
data_merge.head()
```

```python
df = data_merge.copy()
```

```python
## Rename columns
df = df.rename(columns = {'Series or Movie':'series_movie', 'Country Availability':'country','IMDb Score':'IMDb_score','Awards Received':'Awards_received','Awards Nominated For':'Awards_nominated_for','Netflix Release Date':'Netflix_released_date','IMDb Votes':'IMDb_votes','View Rating':'View_rating'})
```

```python
# Convert Date time format
df['Netflix_released_date'] = pd.to_datetime(df['Netflix_released_date'])

# Create year and month columns
df['Released_Year'] = df['Netflix_released_date'].dt.year
df['Released_Year'].astype(int)
df['Released_Month'] = df['Netflix_released_date'].dt.month
df['Released_Month'] = pd.to_datetime(df['Released_Month'], format = '%m').dt.strftime('%b')
```

```python
# Drop irrelevant columns
df.drop(["Title", "Image", "Netflix_released_date", "Languages","Tags", "Runtime", "country", "Hidden Gem Score","Rotten Tomatoes Score","Metacritic Score", "Release Date", "Production House", "Netflix Link","IMDb Link","TMDb Trailer","Trailer Site","Poster"], axis=1, inplace = True)
```

```python
## Remove rows with null values
df.dropna(subset = ['IMDb_votes','Summary','Genre'], inplace = True)
```

```python
# Handling missing values for "categorical" columns
df[['Director','Writer','Actors','View_rating']] = df[['Director','Writer','Actors','View_rating']].fillna(value = "unknown")
```

```python
# Handling missing values for "continuous" columns
df.Awards_received.fillna(0, inplace = True)
df.Awards_nominated_for.fillna(0, inplace = True)
df.Boxoffice.fillna(0, inplace = True)
```

```python
df.isnull().sum() #updated df
```

```python
# Change Column format
df['IMDb_votes'] = df['IMDb_votes'].astype(int)
df['Awards_received'] = df['Awards_received'].astype(int)
df['Awards_nominated_for'] = df['Awards_nominated_for'].astype(int)
df['Boxoffice'] = pd.to_numeric(df['Boxoffice'].str.replace(',', '').str.replace('$', ''))
```

```python
## Create multi-class target column

def categorize(value, thresholds):
    if value <= thresholds[0]:
        return 'Low'
    elif value <= thresholds[1]:
        return 'Medium'
    else:
        return 'High'

score_thresholds = [df['IMDb_score'].quantile(0.33), df['IMDb_score'].quantile(0.67)]
df['IMDb_score_target1'] = df['IMDb_score'].apply(lambda x: categorize(x, score_thresholds))

votes_thresholds = [df['IMDb_votes'].quantile(0.33), df['IMDb_votes'].quantile(0.67)]
df['IMDb_votes_target2'] = df['IMDb_votes'].apply(lambda x: categorize(x, votes_thresholds))

df.head()
```

```python
df1 = df.copy()
```

#### Using both 'Label encoder' and 'One-Hot encoder' to encode the categorical var with numerical value

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

series_movie_le = LabelEncoder()
Director_le = LabelEncoder()
Actors_le = LabelEncoder()
Writer_le = LabelEncoder()
View_rating_le = LabelEncoder()
Released_month_le = LabelEncoder()
IMDb_score_target1_le = LabelEncoder()
IMDb_votes_target2_le = LabelEncoder()
```

```python
df1['series_movie'] = series_movie_le.fit_transform(df1['series_movie'])
df1['Director'] = Director_le.fit_transform(df1['Director'])
df1['Actors'] = Actors_le.fit_transform(df1['Actors'])
df1['Writer'] = Writer_le.fit_transform(df1['Writer'])
df1['View_rating'] = View_rating_le.fit_transform(df1['View_rating'])
df1['Released_Month'] = Released_month_le.fit_transform(df1['Released_Month'])
df1['IMDb_score_target1'] = IMDb_score_target1_le.fit_transform(df1['IMDb_score_target1'])
df1['IMDb_votes_target2'] = IMDb_votes_target2_le.fit_transform(df1['IMDb_votes_target2'])
```

```python
## One hot encoding on "Genre" column
# create a set of unique categories
categories = set()
for row in df1['Genre']:
    for category in row.split(', '):
        categories.add(category)

# create a binary vector for each data point
for category in categories:
  col_name = "Genre_" + category.capitalize()
  df1[col_name] = df1['Genre'].apply(lambda x: 1 if category in x else 0)
```

```python
## Dropping "Genre" columns

df1.drop(['Genre'], axis = 1, inplace = True)
df1.head()
```

```python
def plot_roc_curve(true_y, y_prob):
    """
    plots the roc curve based of the probabilities
    """

    fpr, tpr, thresholds = roc_curve(true_y, y_prob)
    plt.plot([0,1], [0,1],label='Base Rate' 'k--')
    plt.plot(fpr, tpr)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc="lower right")
```
